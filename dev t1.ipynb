```python
# Importing Libraries & Load Data


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Display dataset structure
print("=== Dataset Overview ===")
print(f"Shape: {iris.shape}\n")
print("Columns:", iris.columns.tolist(), "\n")
print("First 5 Rows:\n", iris.head(), "\n")
print("Statistical Summary:\n", iris.describe(), "\n")
```

```python
# Basic Visual Explorations

sns.set(style="whitegrid")  # Professional visual theme

# --- Scatter Plot ---
plt.figure(figsize=(7, 5))
sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette="Set2",
    data=iris,
    s=80,
    edgecolor="black"
)
plt.title("Sepal Length vs Sepal Width (Species-wise)", fontsize=14)
plt.xlabel("Sepal Length", fontsize=12)
plt.ylabel("Sepal Width", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# --- Histogram ---
plt.figure(figsize=(7, 5))
sns.histplot(
    iris["sepal_length"],
    kde=True,
    bins=20,
    edgecolor="black"
)
plt.title("Distribution of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Box Plot Visualization


plt.figure(figsize=(7, 5))
sns.boxplot(
    x="species",
    y="petal_length",
    hue="species",      # Added to satisfy new seaborn rules
    legend=False,       # Prevent duplicate legend
    palette="Set3",
    data=iris
)

plt.title("Petal Length Distribution Across Species", fontsize=14)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length", fontsize=12)
plt.tight_layout()
plt.show()

```
